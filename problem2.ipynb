```
# import libraries
import  numpy as np
from scipy.integrate import odeint 
import matplotlib.pyplot as plt 
```

```
# define model
def model(y, t, k):
  if (t <= 10):
      u = 0.0
  else:
      u = 2.0
  dydt = (-k*y + u)/5
  return dydt
```

```
# initial conditions
y0 = 1
```

```
# time points
t = np.linspace(0,20)
```

```
# solve ODE
k = 1
y = odeint(model,y0,t, args=(k,))
```

```
# plot results
plt.plot(t, y, "r-",linewidth = 2, label = "k=1")
plt.xlabel("time")
plt.ylabel("y(t)")
plt.legend()
plt.show
```
